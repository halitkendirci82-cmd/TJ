```
import pandas as pd
from faker import Faker
import random


fake = Faker('tr_TR')
```

```
class UserProfile:
    """A class to represent a user profile."""
    def __init__(self):
        self.full_name = fake.name()
        self.age = random.randint(18, 75)
        self.address = fake.address().replace("\n", " ")
        self.phone_number = fake.phone_number()
        self.job_title = fake.job()
        self.hometown = fake.city()
        self.salary = random.randrange(25000, 75000, 1000)

    def to_dict(self):
        """Returns the user profile as a dictionary for Pandas."""
        return {
            "Full Name": self.full_name,
            "Age": self.age,
            "Hometown": self.hometown,
            "Address": self.address,
            "Phone Number": self.phone_number,
            "Job Title": self.job_title,
            "Salary (TL)": self.salary
        }
```

```
user_list = [UserProfile() for _ in range(1000)]
df = pd.DataFrame([user.to_dict() for user in user_list])

print("--- Updated Dataset Preview ---")
print(df.head())
```

```
class TJ:
    def __init__(self, name, role, position=0):
        self.name = name
        self.role = role
        self.position = position
        self.target = None

    def set_target(self, target_user):
        self.target = target_user

    def following_mode(self):
        if self.target:

            self.target.position += random.randint(1, 5)


            speed = random.randint(2, 6)
            self.position += speed


            distance = self.target.position - self.position

            print(f"[{self.name}] is following [{self.target.name}]!")
            print(f"-> {self.target.name} Pos: {self.target.position} | {self.name} Pos: {self.position}")
            print(f"-> Distance remaining: {max(0, distance)}\n")

            return distance
```

```

tom = TJ("Tom", "Chaser", position=0)
jerry = TJ("Jerry", "Runner", position=10)

tom.set_target(jerry)

print(f"--- Chase Start: {tom.name} vs {jerry.name} ---\n")

for i in range(1, 11):
    print(f"Round {i}:")
    dist = tom.following_mode()

    if dist <= 0:
        print(f"*** MOUSE TRAPPED! {tom.name} caught {jerry.name}! ***")
        break
```
